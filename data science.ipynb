Visualizations:-

```python
data.corr()
plt.figure(figsize=(15,8))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
plt.figure(figsize=(8,8))
textprops = {"fontsize":18}
plt.title('Category')
plt.pie(data['Category'].value_counts(), labels=data['Category'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```python
plt.figure(figsize=(10,8))
textprops = {"fontsize":18}
plt.title('Region')
plt.pie(data['Region'].value_counts(), labels=data['Region'].value_counts().index,autopct='%1.1f%%',textprops = textprops)
plt.show()
```

```python
plt.figure(figsize=(15,10))
data.groupby('Category')['Profit','Sales','Discount'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(20,10))
data.groupby('Sub-Category')['Profit','Sales','Discount'].agg(['sum']).plot.bar()
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=data['Profit'], y=data['Segment'])
```

```python
sns.pairplot(data)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x=data.State)
plt.xticks(rotation=90)
plt.show()
```
